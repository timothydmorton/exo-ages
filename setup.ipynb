```python
import pandas as pd

df3.rename(columns={'log(g)':'logg', 'vbroad':'Vbroad', 
                    'vmac':'Vmac', 'vsini':'Vsini'},
          inplace=True) # make sure names are consistent
df_all = pd.concat([df3.merge(df4), df8.merge(df9.merge(df11))]) # 3,4 share index; 8,9,11 share index
df_all.to_hdf('data/spec_all.h5', 'df')
```

```python
from configobj import ConfigObj
import os

unc = {'Teff':50, 'feh':0.03, 'logg':0.05}

def write_ini_files(rootdir):
    for _,row in df_all.iterrows():
        name = str(row.Name).replace(' ', '')
        stardir = os.path.join(rootdir, name)
        if not os.path.exists(stardir):
            os.makedirs(stardir)
        
        config = ConfigObj(os.path.join(stardir, 'star.ini'))
        config['Teff'] = (row.Teff, unc['Teff'])
        config['logg'] = (row.logg, unc['logg'])
        config['feh'] = (row['[Fe/H]'], unc['feh'])
        config.write()
```

```python
write_ini_files('starfits')
```
